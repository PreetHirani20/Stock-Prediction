```python
stock_symbols = ['TSLA' , 'AAPL',  'GOOGL']
```

```python
stock_data={}
```

```python
for symbol in stock_symbols:
    
    stock_data[symbol] = pd.read_csv(f'stock_data_{symbol}.csv')
    stock_data[symbol]['Date'] = pd.to_datetime(stock_data[symbol]['Date'])
    stock_data[symbol].sort_values(by='Date', inplace=True)
    stock_data[symbol].set_index('Date', inplace=True)
```

```python
stock_data_GOOGL
```

```python
models = {}
```

```python
for symbol in stock_symbols:
    X = stock_data[symbol][['Close']]
    y = stock_data[symbol]['Close'].shift(-1)
    X_train, X_test, y_train, y_test = train_test_split(X[:-1], y.dropna(), test_size=0.2, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    models[symbol] = model
```

```python
X_train
```

```python
y_train
```

```python
mae_scores = []

for symbol in stock_symbols:
    X_test = stock_data[symbol][['Close']].iloc[-1:]
    y_test = stock_data[symbol]['Close'].iloc[-1]
    
    y_pred = models[symbol].predict(X_test)
    
    # Calculate MAE
    mae = mean_absolute_error([y_test], y_pred)
        
    # Append scores to the lists
    mae_scores.append(mae)
    
    print(f"Stock Symbol: {symbol}")
    print(f"True Price: {y_test}")
    print(f"Predicted Price: {y_pred[0]}")
    print(f"Mean Absolute Error (MAE): {mae}")
    print()

# # Calculate and print the average scores for all symbols
# avg_mae = sum(mae_scores) / len(mae_scores)

# print("Average Mean Absolute Error (MAE): ", avg_mae)
```

```python
y_test
```

```python
n_steps = 5

# Make predictions for the next n_steps
y_pred = []
for i in range(n_steps):
    X_test = stock_data['TSLA'][['Close']].iloc[-n_steps:]
    next_pred = models['TSLA'].predict(X_test.iloc[i:i+1])
    y_pred.append(next_pred[0])

# Visualize the predictions for TSLA.
plt.figure(figsize=(10, 5))
plt.title(f"Stock Price Prediction for TSLA (Next {n_steps} Steps)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.plot(stock_data['TSLA'].index[-n_steps:], X_test, label="Actual Prices", color='blue')
plt.plot(stock_data['TSLA'].index[-n_steps:], y_pred, label="Predicted Prices", color='red')
plt.legend()
plt.show()
```
